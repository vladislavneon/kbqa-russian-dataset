```python
%config IPCompleter.greedy=True
import re
import json
import pandas as pd
from functools import partial
from collections import defaultdict
from sklearn.feature_extraction.text import CountVectorizer
from tqdm import tqdm_notebook as tqdm
```

Read questions

```python
cv = CountVectorizer(token_pattern=r'\w+')
analyzer = cv.build_analyzer()
```

```python
quiz_q = {}
with open('quiz_full.txt', 'r') as inf:
    que = None
    ans = None
    for i, line in enumerate(inf):
        if i % 2 == 0:
            que = line.strip()
        else:
            ans = line.strip()
            quiz_q[' '.join(analyzer(que))] = (que, ans)
```

```python
full_q = list(quiz_q.keys())
```

```python
# %load pattern_search.py
def finder_substr(q, prop):
    if re.search(prop, q) is not None:
        return True
    return False

def apply_properties_finder(q, finder, prop):
    f = partial(finder, prop=prop)
    return list(filter(f, q))

def find_property(aliases):
    qs = []
    for alias in aliases:
        cur_qs = apply_properties_finder(full_q, finder_substr, alias)
        qs.extend(cur_qs)
    return qs
```

```python
with open('properties_aliases.json', 'r') as inf:
    prop_aliases = json.load(inf)
    
with open('pids.json', 'r') as inf:
    pids = json.load(inf)
```

```python
retrieved_table = []
with open('retrieved_quiz_forms.txt', 'w') as ouf:
    pass
with open('retrieved_quiz_forms.txt', 'a') as ouf:
    for pid in tqdm(pids[:100]):
        print(pid[0], file=ouf)
        print(pid[1], file=ouf)
        print(prop_aliases[pid[0]], file=ouf)
        for q in find_property(prop_aliases[pid[0]]):
            print(q, file=ouf)
            row = {}
            row['pid'] = pid[0]
            row['label'] = pid[1]
            row['q'] = q
            row['good'] = ''
            retrieved_table.append(row)
        print(file=ouf)

retrieved_df = pd.DataFrame(retrieved_table)[['pid', 'label', 'q', 'good']]
```

### Question words

```python
question_words = []
with open('question_words.txt', 'r') as inf:
    for line in inf:
        question_words.append(line.strip('\n'))
```

```python
def starts_with_question_word(q):
    for qw in question_words:
        if re.match(qw, q) is not None:
            return True
    return False

def contains_not(q):
    return ' не ' in q

def too_short(q):
    if len(q.split()) < 4:
        return True
    return False

stop_substrs = [
    ' означ',
    'в народе',
    'на руси',
    'в просто',
    'что такое',
    'кто такой',
    'кто такая'
]
def contains_stop_substr(q):
    for sb in stop_substrs:
        if sb in q:
            return True
    return False
```

```python
def check_filters(q):
    if not too_short(q) and \
       not contains_not(q) and \
       not contains_stop_substr(q) and \
       starts_with_question_word(q):
        return True
    return False
```

```python
filtered_q = []
with open('quiz_filtered.txt', 'w') as ouf:
    cnt = 0
    for q in tqdm(full_q):
        if check_filters(q):
            filtered_q.append(q)
            ouf.write(q + '\n')
            cnt += 1
    print(cnt)
```
